```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/shinten/Desktop/space_econometrics/figures/data/iQPS_data.csv')
```

```python
# Total Assets, Liabilities, Net IncomeをYearに対してプロット
ax = df.plot(x='Year', y=['Total assets', 'Total Liabilities', 'Total Revenue', 'Capital Surplus', 'Net Income'], title='Year vs Variables')

# X軸とY軸のラベルを設定 axは、Matplotlibライブラリを使用してグラフを作成する際によく使われる変数です。これは「Axes」オブジェクトの略称であり、グラフの個々の領域を表します。Axesオブジェクトには、グラフのX軸やY軸のラベルを設定するメソッドや、タイトルを設定するメソッド、データをプロットするメソッドなど、グラフの見た目やデータ表示に関する多くの機能が含まれています。
ax.set_xlabel('Year')
ax.set_ylabel('Values')

# 凡例を表示（plotメソッドのlegendパラメータがTrueの場合は自動的に表示されます）
plt.legend()

# グラフの表示
plt.show()
```

```python
# 'Year'をX軸に、複数の変数をY軸に設定してプロット
ax = df.plot(x='Year', y=['Total assets', 'Total Liabilities', 'Total Revenue', 'Capital Surplus', 'Net Income'], title='Year vs Variables')

# X軸とY軸のラベルを設定
ax.set_xlabel('Year')
ax.set_ylabel('Values')

# 凡例を表示
plt.legend()

# グラフの表示
plt.show()
```
